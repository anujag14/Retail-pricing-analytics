```python
# UK Supermarket Data Preprocessing
This notebook loads, cleans, and prepares supermarket datasets for machine learning and time series analysis.
```

```python
import pandas as pd
import numpy as np
import os
```

```python
path = r"D:\uk-supermarket-forecasting\data\raw"
files = os.listdir(path)

dfs = []

for file in files:
    if file.endswith(".csv"):   # ✅ FIX HERE
        df = pd.read_csv(os.path.join(path, file))
        df["supermarket"] = file.split(".")[0]
        dfs.append(df)

# Combine
df_all = pd.concat(dfs, ignore_index=True)

print("Data loaded successfully")
df_all.head()
```

```python
df_all["date"] = pd.to_datetime(df_all["date"], errors="coerce")
```

```python
df_all = df_all.dropna()
```

```python
df_all["year"] = df_all["date"].dt.year
df_all["month"] = df_all["date"].dt.month
df_all["day"] = df_all["date"].dt.day
df_all["dayofweek"] = df_all["date"].dt.dayofweek
```
